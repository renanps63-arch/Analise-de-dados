## Problem 4

##### Enumerar todos os crimes de sequestro envolvendo crianças.

### Problema 5

##### Listar os tipos de crimes registrados nas escolas. (Sem repetições)

```python
%%sql select distinct primary_type from chicago_crime_data
where location_description like '%school%';
```

### Problema 6

##### Listar os tipos de escolas, juntamente com a média de segurança de cada tipo.

```python
%%sql select "Elementary, Middle, or High School", AVG(safety_score) AS "AVG_Safety_Score"
from CHICAGO_PUBLIC_SCHOOLS
group by "Elementary, Middle, or High School";
```

```python
# Solução alternativa devido problemas da extensão ipython-sql com nomes de colunas que contém vígulas.
query = '''
SELECT "Elementary, Middle, or High School", AVG(safety_score) AS "AVG_Safety_Score"
FROM CHICAGO_PUBLIC_SCHOOLS
GROUP BY "Elementary, Middle, or High School"
'''
df = pd.read_sql_query(query, con)
df.head()
```

### Problem 7

##### List 5 community areas with highest % of households below poverty line

```python
%%sql select community_area_name from Census_data
order by PERCENT_HOUSEHOLDS_BELOW_POVERTY desc
limit 5;
```

### Problema 8

##### Qual área comunitária é mais propensa ao crime? Indique apenas o número da área comunitária.

```python
%%sql SELECT COMMUNITY_AREA_NUMBER
FROM CHICAGO_CRIME_DATA
WHERE COMMUNITY_AREA_NUMBER IS NOT NULL
GROUP BY COMMUNITY_AREA_NUMBER
ORDER BY COUNT(*) DESC
    limit 1;
```

### Problema 9

##### Use uma subconsulta para encontrar o nome da área comunitária com o maior índice de vulnerabilidade.

```python
%%sql select community_area_name from Census_data
where hardship_index IN (select max(hardship_index) from Census_data);
```

### Problema 10

##### Use uma subconsulta para determinar o nome da área comunitária com o maior número de crimes.

```python
%%sql SELECT COMMUNITY_AREA_NAME
FROM CENSUS_DATA
WHERE COMMUNITY_AREA_NUMBER = 
    (SELECT COMMUNITY_AREA_NUMBER FROM CHICAGO_CRIME_DATA
    WHERE COMMUNITY_AREA_NUMBER IS NOT NULL
    GROUP BY COMMUNITY_AREA_NUMBER
    ORDER BY COUNT(*) DESC
    LIMIT 1);
```
